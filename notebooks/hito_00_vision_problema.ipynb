### 1.1 El Escenario de partida

Se va acercando la fecha del verano y la gente se va de vacaciones y la principal forma que tienen muchas personas de disfrutar al máximo estas vacaciones es alquilando un piso, casa, chalet, habitaciones de hotel o variantes. Yo como cliente me he pasado horas y horas bicheando, comparando e investigando en diferentes páginas como trivago o airBNB. El resultado suele ser el mismo: decenas de pestañas abiertas, parálisis por análisis y una sensación de desconcierto absoluto ante precios desorbitados, tanto al alza como a la baja.

Sin embargo el problema que viene a acotar este proyecto no se enfoca en el lado de los cliente sino al contrario. En la actualidad, la fijación de precios en las plataformas de alquiler vacacional adolece de un problema endémico: **la rigidez o la heurística intuitiva**. Tradicionalmente, la gran mayoría de los propietarios o pequeños gestores de activos determinan sus tarifas basándose en tres metodologías subóptimas:

* **Intuición empírica ("gut feeling"):** Fijación de precios basada en percepciones sesgadas del valor de su propia propiedad.
* **Medias históricas estáticas:** Replicar las tarifas del año anterior aplicando un porcentaje plano de inflación, ignorando los microcambios macroeconómicos del presente.
* **Estrategias espejo pasivas:** Copiar a ciegas los precios de los competidores más cercanos, asumiendo que sus costes de estructura y sus tasas de ocupación son idénticos.

Esta desconexión con la realidad analítica del mercado genera **ineficiencias operacionales críticas**:

* **El coste de la sobrevaloración (Overpricing):** Establecer un precio por encima del umbral de tolerancia del mercado resulta en ventanas de vaciado prolongadas. Un inmueble vacío no solo deja de facturar ingresos directos, sino que sigue devengando costes fijos (mantenimiento, suministros, amortizaciones), destruyendo el margen de beneficio neto de la explotación.
* **El coste de oportunidad por infravaloración (Underpricing):** Establecer precios por debajo de la disposición máxima a pagar del cliente (*willingness to pay*), especialmente en momentos de picos de demanda (festivos, macroeventos locales, ferias internacionales), se traduce en "dejar dinero sobre la mesa". El alojamiento se reserva con demasiada antelación, perdiendo la capacidad de capturar el valor real de esa ventana temporal.

### 1.2 La Propuesta de Valor: Optimización Avanzada del RevPAR

La utilidad real de este proyecto no radica únicamente en la construcción de un estimador matemático, sino en el diseño e implementación de un **Sistema de Precios Dinámicos Inteligente**. La métrica reina que este sistema busca maximizar es el **RevPAR (Revenue per Available Room)**, el indicador de rendimiento estándar en la industria turística, cuya fórmula conceptual se define como:

$$\text{RevPAR} = \text{Tasa de Ocupación} \times \text{Precio Medio Diario (ADR)}$$

Fijar el precio óptimo exige romper con el paradigma de las bases de datos aisladas. Para maximizar el RevPAR, el sistema propuesto unifica **tres dimensiones críticas del mercado que operan a velocidades y con estructuras radicalmente distintas**, consolidando una visión de 360 grados sobre el ecosistema del inmueble:

* **1. La Infraestructura Física (Dimensión Estática/Estructurada):** Representa las características intrínsecas e inmutables del activo inmobiliario (tipología de propiedad, capacidad, número de habitaciones, camas y baños). Esta información, gestionada mediante sistemas relacionales tradicionales (**Amazon RDS**), define el *precio suelo o umbral mínimo de rentabilidad*, el límite por debajo del cual la explotación del inmueble deja de ser viable.

* **2. La Reputación Digital y Calidad Percibida (Dimensión Semiestructurada/Textual):** Las métricas de valoración tradicionales (las clásicas puntuaciones de 1 a 5 estrellas) suelen sufrir inflación y ofrecen una visión limitada. La verdadera ventaja competitiva y el valor diferencial percibido por el cliente residen en el texto libre de las reseñas. Mediante técnicas de **Procesamiento de Lenguaje Natural (NLP)** sobre almacenes documentales (**MongoDB Atlas**), el sistema extrae un *score de sentimiento* que cuantifica factores intangibles (hospitalidad, nivel de ruido, limpieza real o comodidad del check-in). Esto permite aplicar un *premium* (sobreprecio) justificado por la calidad de la experiencia del usuario.

* **3. El Pulso del Mercado en Tiempo Real (Dimensión Dinámica/Streaming):** La demanda en el sector es extremadamente líquida; varía minuto a minuto según el volumen de usuarios concurrentes buscando alojamiento en una zona geográfica concreta. Capturar este comportamiento antes de que se formalice la reserva resulta vital. Utilizando una arquitectura de eventos distribuidos (**Apache Kafka**), el sistema procesa la telemetría web (clics, búsquedas, visualizaciones) para generar un índice de presión de la demanda, lo que fuerza al modelo predictivo a ajustar las tarifas de manera elástica ante picos repentinos de interés.

> **Nota metodológica sobre la ingesta de datos en tiempo real:** Dado que la obtención de metadatos de telemetría web en plataformas de terceros (tráfico en vivo) es restringida, esta tercera dimensión se abordará mediante la simulación de demanda. Para ello, se desarrollará un generador de tráfico sintético (*script* de simulación) que emulará el volumen de usuarios concurrentes interactuando con los alojamientos. ## 2. Definición del Problema de Machine Learning

Este proyecto se enmarca estrictamente dentro del paradigma del Aprendizaje Supervisado (Supervised Learning). La justificación de esta aproximación radica en que el algoritmo se entrenará utilizando un conjunto de datos etiquetado, donde cada registro histórico ya cuenta con la respuesta correcta asociada: el precio real de mercado.

Dentro del aprendizaje supervisado, el sistema se define como un problema de Regresión Multivariable, dado que la variable que se desea determinar es numérica, continua y depende de la interacción simultánea de múltiples factores predictivos. No se trata de un problema de clasificación (ya que no se busca categorizar los inmuebles en clases discretas), ni de aprendizaje no supervisado (puesto que el objetivo no es descubrir agrupaciones u ordenamientos ocultos, sino modelar una función matemática de costes y tarifas).

## 3. Definición de Variable Objetivo (Target) y Variables Independientes (Features)

Para garantizar la viabilidad técnica del modelo, se delimitan formalmente los componentes que interactuarán en el algoritmo predictivo:

* **Variable Objetivo o Target ($Y$):** Es la variable dependiente que el modelo debe estimar de forma óptima. En este dataset está representada por la columna **`price`** (convertida a tipo numérico continuo *Float* tras un proceso de limpieza de caracteres monetarios). Su fin es determinar la tarifa por noche ideal para maximizar el RevPAR.
* **Variables Independientes o Features ($X$):** Son el conjunto de características predictivas extraídas de las bases de datos (RDS y MongoDB) y del flujo de eventos (Kafka) que el modelo utilizará para inferir el valor de la variable objetivo.

### 3.1 Identificación y Justificación de las Variables Relevantes

No todas las columnas presentes en el dataset original aportan valor predictivo; introducir variables irrelevantes (como `host_thumbnail_url` o `scrape_id`) introduciría ruido y provocaría sobreajuste (*overfitting*). A continuación, se justifican las variables seleccionadas, clasificadas según la dimensión de negocio que representan:

#### 3.1.1 Dimensión Espacial y de Ubicación Geográfica

La localización es el factor más determinante en el sector inmobiliario y turístico, ya que define el acceso a servicios, transportes y puntos de interés.

* **`latitude` y `longitude` (Float):** Permiten al modelo calcular distancias exactas a puntos críticos de la ciudad (monumentos, playas o centros financieros) y modelar vecindarios de precios por cercanía geográfica real, superando las limitaciones de las fronteras políticas de los barrios.
* **`neighbourhood_cleansed` (Categorizada):** Aporta el contexto socioeconómico y de demanda del barrio donde se ubica el inmueble, permitiendo al algoritmo aplicar coeficientes base según la exclusividad de la zona.

#### 3.1.2 Dimensión de Infraestructura Física (Atributos del Inmueble)

Estas variables definen la capacidad operativa de la propiedad y marcan el *precio suelo* (el coste mínimo de explotación por sábanas, limpieza y suministros por debajo del cual el propietario incurre en pérdidas).

* **`room_type` (Categorizada):** Modifica radicalmente el valor percibido. Un apartamento completo (`Entire home/apt`) cuenta con una disposición a pagar muy superior por parte del cliente frente a una habitación privada (`Private room`) o compartida.
* **`accommodates` (Int):** Define la capacidad máxima de huéspedes. Existe una correlación lineal directa entre el número de personas que pueden alojarse y el precio total de la estancia.
* **`bedrooms` y `beds` (Int):** Determinan la comodidad y la segmentación del cliente (viajeros solitarios, parejas o familias/grupos grandes), lo que impacta directamente en la elasticidad del precio.
* **`bathrooms_text` (String/Transformada):** La proporción de baños por huésped es un indicador crítico de calidad. Se procesará este campo de texto para extraer el número neto de baños disponibles.

#### 3.1.3 Dimensión de Reputación Digital y Calidad Percibida

El comportamiento del consumidor turístico está guiado por la prueba social y la confianza en la plataforma. Un alojamiento con excelentes reseñas reduce la incertidumbre del cliente, lo que justifica la aplicación de un precio *premium*.

* **`review_scores_rating` (Float):** Funciona como el indicador macro de la satisfacción general del histórico de clientes.
* **`review_scores_cleanliness` y `review_scores_location` (Float):** Permiten ponderar el precio en función de dos de las mayores preocupaciones del huésped: la higiene real del sitio y la veracidad de la ubicación descrita.
* **`score_sentimiento_nlp` (Float - Variable derivada):** Variable continua generada a través del análisis de sentimiento con NLP sobre el archivo complementario `reviews.csv` (almacenado en MongoDB Atlas). Captura matices cualitativos en un rango de $[-1, 1]$ (ruido, comodidad del colchón, amabilidad del anfitrión) que las notas numéricas tradicionales no reflejan.

#### 3.1.4 Dimensión de Presión de la Demanda (Tiempo Real)

* **`volumen_clicks_15min` (Int - Variable sintética en streaming):** Calculada mediante la agregación de eventos en tiempo real provenientes de Apache Kafka. Es la variable elástica del modelo: si un alojamiento experimenta un pico inusual de accesos ("usuarios entrando a ver la publicación") en la última ventana temporal de 15 minutos, esta variable forzará al algoritmo a elevar la tarifa para capturar ese coste de oportunidad antes de que se agote el inventario de la zona.

## 4. Integración Multifuente e Información Híbrida

La innovación técnica de este pipeline radica en que integra datos con tres estructuras y velocidades de ingesta completamente diferentes:
1. **Estructurada (SQL/RDS):** Almacena el inventario maestro de las propiedades.
2. **Semiestructurada (NoSQL/MongoDB Atlas):** Alberga las reseñas masivas en documentos JSON flexibles.
3. **Streaming en Tiempo Real (Apache Kafka):** Captura el flujo constante de búsquedas antes de que se consoliden en reservas.

A continuación, se presenta la celda de código que ejecuta la simulación de integración híbrida utilizando **Pandas** para consolidar las tres fuentes en una única 'Súper Tabla' analítica que será persistida en el Data Lake de **Amazon S3** (Esto es un fragmeto del código [pipeline_inicio.py](../scripts/pipeline_inicio.py)).

````python
import pandas as pd
import json

print("--- INICIANDO PROCESO DE INTEGRACIÓN MULTIFUENTE HITO 0/1 ---")

# 1. Simulación de datos físicos desde la base de datos relacional (RDS)
rds_listings = [
    {"listing_id": 101, "property_type": "Apartamento", "accommodates": 4, "precio_noche": 120.0},
    {"listing_id": 102, "property_type": "Estudio", "accommodates": 2, "precio_noche": 85.0}
]
df_rds = pd.DataFrame(rds_listings)

# 2. Simulación de opiniones cualitativas desde MongoDB Atlas
mongo_reviews = [
    {"listing_id": 101, "comments": "Excelente ubicación y muy limpio.", "score_sentimiento": 0.92},
    {"listing_id": 102, "comments": "Un poco ruidoso por las noches.", "score_sentimiento": 0.51}
]
df_mongo = pd.DataFrame(mongo_reviews)

# 3. Simulación de clics de demanda en tiempo real desde Apache Kafka
kafka_stream = [
    {"listing_id": 101, "ratio_busquedas_zona": 1.4},
    {"listing_id": 101, "ratio_busquedas_zona": 1.6},
    {"listing_id": 102, "ratio_busquedas_zona": 0.9}
]
df_kafka = pd.DataFrame(kafka_stream)

# Agregación estadística de las fuentes dinámicas antes del acoplamiento
df_kafka_agg = df_kafka.groupby('listing_id')['ratio_busquedas_zona'].mean().reset_index()

# JOIN Lógico utilizando 'listing_id' como llave de unificación
df_master = df_rds.merge(df_mongo[['listing_id', 'score_sentimiento']], on='listing_id', how='left')
df_master = df_master.merge(df_kafka_agg, on='listing_id', how='left')

print("\n✔ ¡Dataset unificado generado con éxito para la fase de Machine Learning!")
print("Estructura final del Data Lake (Muestra):")
df_master.head()
````
