# Hito 1: Visión del Problema y Arquitectura de Datos para un Sistema PropTech de Precios Dinámicos

## 1. Definición formal del problema

El proyecto se enmarca en un problema de **aprendizaje supervisado** de tipo **regresión**. El objetivo es construir un modelo capaz de estimar el **precio óptimo por noche** (`precio_noche`) para alquileres vacacionales, a partir de un conjunto de variables explicativas que capturan características del inmueble, la calidad de las reseñas y la actividad de demanda en tiempo real.

### 1.1. Definición matemática

Dados un conjunto de registros `X = {x_1, x_2, ..., x_n}` y una variable objetivo `y`, buscamos una función `f(X)` tal que:

`precio_noche = f(property_type, accommodates, bathrooms, bedrooms, beds, comments, score_sentimiento, ratio_busquedas_zona) + ε`

donde `ε` representa el error aleatorio. El modelo debe aprender de ejemplos históricos para generalizar a nuevos anuncios.

### 1.2. Contexto de utilidad real

En el sector PropTech, las plataformas que gestionan alquileres vacacionales requieren decisiones de precios automatizadas que maximicen ocupación y rentabilidad sin sacrificar competitividad. Un sistema de precios dinámicos aporta valor real al: 

- Propietario o gestor, que obtiene ingresos más altos y reduce noches sin ocupar.
- Huésped, recibiendo precios más adecuados al mercado y a la demanda.
